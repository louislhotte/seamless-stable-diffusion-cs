```python
def create_mask(image):
    window_name = "Image"
    # Initialize the mask
    mask = np.zeros(image.shape[:2], dtype=np.uint8)
    resized_image = image.copy()
    resized_mask = mask.copy()
    zoom_level = 0
    last_zoom_positions_level = (0,0,0)

    # Mouse callback function to draw on the mask
    radius = 10

    def resizing_image(image_, zoom_factor, zoom_level, x, y):
        # Get the dimensions of the image
        h, w = image_.shape[:2]

        # Calculate the new dimensions
        new_h, new_w = int(h * (1+zoom_level)), int(w * (1+zoom_level))

        # Resize the image
        if zoom_factor>1:
            resized_image_local = cv2.resize(image_, (new_w, new_h), interpolation=cv2.INTER_LINEAR)
        else:
            resized_image_local = cv2.resize(image_, (new_w, new_h), interpolation=cv2.INTER_AREA)

        # Calculate the region to crop to keep the cursor position in the same place
        x1 = max(0, int(x * (1+zoom_level) - x))
        y1 = max(0, int(y * (1+zoom_level) - y))
        x2 = min(new_w, x1 + w)
        y2 = min(new_h, y1 + h)

        # Crop the resized image
        resized_image_local = resized_image_local[y1:y2, x1:x2]
        return(resized_image_local)
    
    def get_zoom_position_level(x,y,zoom_level):
        x1 = max(0, int(x - (x / (1+zoom_level))))
        y1 = max(0, int(y - (y / (1+zoom_level))))
        print("Zoom position (x1,y1): ", x1,y1)
        return((x1,y1,zoom_level))
    
    def get_origin_xy(x,y):
        x1,y1,zoom_level_ = last_zoom_positions_level
        orig_x = int(x1 + x / (1+zoom_level))
        orig_y = int(y1 + y / (1+zoom_level))
        print("Origin (x,y,orig_x,orig_y): ",(x, y, orig_x, orig_y))
        return(orig_x,orig_y)

    def draw_mask(event, x, y, flags, param):
        nonlocal radius, resized_image, resized_mask, zoom_level, last_zoom_positions_level  # Use the nonlocal keyword to modify the outer scope variable
        if event == cv2.EVENT_LBUTTONDOWN or (event == cv2.EVENT_MOUSEMOVE and flags & cv2.EVENT_FLAG_LBUTTON):
            cv2.circle(resized_mask, (x, y), radius, 255, -1)  # Draw a circle on the mask
            print("x,y:", x,y)
            
            # Calculate the coordinates on the original mask
            orig_x,orig_y = get_origin_xy(x,y)
            orig_radius = int(radius / (1 + zoom_level))
            cv2.circle(mask, (orig_x, orig_y), orig_radius, 255, -1)

        elif event == cv2.EVENT_MOUSEWHEEL:
            zoom_factor = 1
            if flags > 0:
                zoom_factor = 1.1  # Zoom in
            elif zoom_level>0:
                zoom_factor = 0.9  # Zoom out
            
            if zoom_factor!=1:
                zoom_level = round(10*(zoom_level + zoom_factor-1))/10
                resized_image = resizing_image(image,zoom_factor, zoom_level, x, y)
                resized_mask  = resizing_image(mask,zoom_factor, zoom_level, x, y)
                last_zoom_positions_level = (get_zoom_position_level(x,y,zoom_level))


    # Display the image and set the mouse callback
    cv2.namedWindow(window_name)
    cv2.setMouseCallback(window_name, draw_mask)

    while True:
        # Create a copy of the image to display the overlay
        overlay = resized_image.copy()
        overlay[resized_mask == 255] = [255, 255, 255]  # Overlay the mask on the image
        
        # Show the image with the mask overlay
        cv2.imshow(window_name, overlay)
        
        # Break the loop when 'q' is pressed or the window is closed
        if (cv2.waitKey(1) & 0xFF == ord('q')) or (cv2.getWindowProperty(window_name, cv2.WND_PROP_VISIBLE) < 1):
            break

    cv2.destroyAllWindows()
    return(mask)


try:
    cv2.imshow("Mask", create_mask(image))
    cv2.waitKey(0)
    cv2.destroyAllWindows()
except Exception as e:
    print(f"An error occurred: {e}")
    cv2.destroyAllWindows()
    
```

```python
def create_mask(image):
    window_name = "Image"
    # Initialize the mask
    mask = np.zeros(image.shape[:2], dtype=np.uint8)

    # Mouse callback function to draw on the mask
    radius = 1
    def draw_mask(event, x, y, flags, param):
        nonlocal radius
        if event == cv2.EVENT_LBUTTONDOWN or (event == cv2.EVENT_MOUSEMOVE and flags & cv2.EVENT_FLAG_LBUTTON):
            cv2.circle(mask, (x, y), radius, 255, -1)  # Draw a circle on the mask
        elif event == cv2.EVENT_MOUSEWHEEL:
            if flags > 0:
                radius += 1  # Increase radius
            else:
                radius = max(1, radius - 1)  # Decrease radius but keep it at least 1

    # Display the image and set the mouse callback
    cv2.namedWindow(window_name)
    cv2.setMouseCallback(window_name, draw_mask)

    while True:
        # Create a copy of the image to display the overlay
        overlay = image.copy()
        overlay[mask == 255] = [255, 255, 255]  # Overlay the mask on the image
        
        # Show the image with the mask overlay
        cv2.imshow(window_name, overlay)
        
        # Break the loop when 'q' is pressed or the window is closed
        if (cv2.waitKey(1) & 0xFF == ord('q')) or (cv2.getWindowProperty(window_name, cv2.WND_PROP_VISIBLE) < 1):
            break

    cv2.destroyAllWindows()
    return(mask)

try:
    cv2.imshow("Mask", create_mask(image))
    cv2.waitKey(0)
    cv2.destroyAllWindows()
except Exception as e:
    print(f"An error occurred: {e}")
    cv2.destroyAllWindows()
```
